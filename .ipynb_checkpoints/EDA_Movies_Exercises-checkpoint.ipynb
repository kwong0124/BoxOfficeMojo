Plot the Domestic Total Gross by Release Date. Try a scatterplot and a line plot. In what scale are the numbers on the Y axis?

```python
%matplotlib inline

x = mv_df['reldate']
y = mv_df['DomesticTotalGross'] / 1000000

fig = plt.figure(figsize=(15,5))
plt.plot(x, y, 'bx')
plt.title('2013-2016 Movies: Domestic Gross Total by Release Date')
plt.xlabel('Date')
plt.ylabel('Domestic Gross Total ($ in Millions)')
plt.suptitle('Exercise 1 - Scatter Plot')
plt.show()
```

```python
fig = plt.figure(figsize=(15,5))
ax = plt.gca() # why is this needed? 
mv_df = mv_df.sort_values(by=['reldate']) # values need to be sorted otherwise the line plot may
# not be in the correct order 
mv_df.plot(x='reldate', y='DomesticTotalGross', ax=ax)
plt.title('2013-2016 Movies: Domestic Gross Total by Release Date')
plt.xlabel('Date')
plt.ylabel('Domestic Gross Total ($ in Billions)')
plt.suptitle('Exercise 1 - Line Plot')
plt.show()
```

### Exercise 3

The Motion Picture Association of America (MPAA) rating system comprises of G (general audiences: all ages admitted), PG (parental guidance suggested: may not be suitable for children), PG-13 (parents stronly cautioned: may be inappropriate for children under 13), and R (restricted: under 17 requires accompanying parent or adult guardian). Find the average Runtime and Domestic Total Gross at each Rating Level. Plot both by Rating. Do you see any pattern in the data?

```python
mv_df.groupby('Rating').count()
```

```python
mv_df.groupby('Rating').mean()[['Runtime', 'DomesticTotalGross']]
```

```python
sorted_by_rating = mv_df.groupby('Rating').mean()[['Runtime', 'DomesticTotalGross']]
```

```python
avg_runtimes = []

for item in sorted_by_rating['Runtime']:
    avg_runtimes.append(item)
```

```python
mean_DGT = []

for item in sorted_by_rating['DomesticTotalGross']:
    mean_DGT.append(item)
```

```python
ratings = ['G', 'PG', 'PG-13', 'R']
x = [1, 2, 3, 4] # locations for the x groups
p1 = plt.bar(x, avg_runtimes, width=0.4, color='b')
plt.xticks(x, ratings)
plt.ylabel('Runtime (Minutes)')
```

```python
x = [1, 2, 3, 4] # locations for the x groups
y = mean_DGT
p1 = plt.bar(x, y, width=0.4, color='r')
plt.xticks(x, ratings)
plt.ylabel('Domestic Total Gross ($)')
```

### Exercise 4

Plot the Domestic Total Gross by the Release Date. Segment by Rating - that is, have all 4 groups on the same plot. Can you spot anything out of the ordinary? Now make 4 separate plots (one for each Rating) but part of the same matplotlib figure. What are the benefits and liabilities of each approach?

```python
mv_df.loc[mv_df['Rating'] == 'G']
```

```python
G_titles = mv_df.loc[mv_df['Rating'] == 'G'] 
PG_titles = mv_df.loc[mv_df['Rating'] == 'PG']
PG_13_titles = mv_df.loc[mv_df['Rating'] == 'PG-13'] 
R_titles = mv_df.loc[mv_df['Rating'] == 'R'] 
```

```python
%matplotlib inline

x_G = G_titles['reldate']
Y_G = G_titles['DomesticTotalGross'] / 1000000

X_PG = PG_titles['reldate']
Y_PG = PG_titles['DomesticTotalGross'] / 1000000

X_PG_13 = PG_13_titles['reldate']
Y_PG_13 = PG_13_titles['DomesticTotalGross'] / 1000000

X_R = R_titles['reldate']
Y_R = R_titles['DomesticTotalGross'] / 1000000

fig = plt.figure(figsize=(15,5))
plt.plot(x_G, y_G, 'bx')
plt.plot(X_PG, Y_PG, 'rx')
plt.plot(X_PG_13, Y_PG_13, 'gx')
plt.plot(X_R, Y_R, 'yx')

plt.title('2013-2016 Movies: Domestic Gross Total by Release Date')
plt.xlabel('Date')
plt.ylabel('Domestic Gross Total ($ in Millions)')
plt.suptitle('Exercise 4')
plt.legend(labels=['G', 'PG', 'PG-13', 'R'])
plt.show()
```

```python
%matplotlib inline

fig = plt.figure(figsize=(20, 15))
plt.subplot(2, 2, 1)
plt.plot(x_G, y_G, 'bx')
plt.title('G')

plt.subplot(2, 2, 2)
plt.plot(X_PG, Y_PG, 'rx')
plt.title('PG')

plt.subplot(2, 2, 3)
plt.plot(X_PG_13, Y_PG_13, 'gx')
plt.title('PG-13')

plt.subplot(2, 2, 4)
plt.plot(X_R, Y_R, 'yx')
plt.title('R')

plt.title('2013-2016 Movies: Domestic Gross Total by Release Date')
plt.xlabel('Date')
plt.ylabel('Domestic Gross Total ($ in Millions)')
plt.suptitle('Exercise 4')
plt.legend(labels=['G', 'PG', 'PG-13', 'R'])
plt.show()
```
